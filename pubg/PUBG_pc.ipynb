```python
df['headshotKills'].value_counts()
```

```python
df = df.drop(['headshotKills'], axis = 1)
```

```python
cols = ['boosts', 'damageDealt', 'DBNOs', 'heals', 'killPlace','killPoints', 'kills', 'killStreaks', 'longestKill', 'matchDuration','rankPoints', 'weaponsAcquired','winPoints', 'winPlacePerc', 'distance']
```

```python
for i in cols:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    print(IQR)
    upper = df[i] >= (Q3+1.5*IQR)
    lower = df[i] <= (Q1-1.5*IQR)
    for j in list(np.where(upper)[0]):
        df[i].replace({df[i][j]},df[i].quantile(0.95), inplace = True)
    for k in list(np.where(lower)[0]):
        df[i].replace({df[i][k]},df[i].quantile(0.05), inplace = True)
```

```python
# Encoding categorical data
df = pd.get_dummies(df,drop_first =  True)
```

```python
df.columns
```

# Training the model

```python
X = df.iloc[:,np.r_[0:15,16:]].values
y = df.iloc[:, 15].values
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
print(y_test)
```

```python
y_pred = regressor.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X, y)
```

```python
y_pred = regressor.predict(X_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
