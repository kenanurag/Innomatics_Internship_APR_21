# The dataset

The given dataset on PUBG contains 1111742 and 30 columns. Let us discuss in brief what the columns are about.

**ID :** This is the player's ID. Each player has a unique ID.\
**Group ID :** The players can play the game solo, in a pair(duo) or in a squad(4 members). The groupID column is the unique ID of a particular group.\
**Match ID :** Each match has 100 players and all these 100 players will have a unique matchID.\
**Match type :** There can be different match modes like solo TPP (Third person perspective),solo FPP (First person perspective) etc.\
The numerical columns contain various aspects of the game like damage dealt, kills , distance travelled etc. The column **winPlacePerc**  is a percentile winning placement, where 1 corresponds to 1st place, and 0 corresponds to last place in the match. This is calculated based on the last placed group in the match,whose data we have available.

# Bivariate analysis

Let us plot the analysis of numerical columns vs winPlacePerc

```python
df.plot.scatter(x = 'damageDealt' , y = 'winPlacePerc')
plt.show()
```

```python
df.plot.hexbin(x = 'damageDealt' , y = 'winPlacePerc')
plt.show()
```

The above plots are rather unappealing. Let us impute the outliers with median and see the result

```python
print(df['damageDealt'].median())
```

```python
median = df['damageDealt'].median()
Q1 = df['damageDealt'].quantile(0.25)
Q3 = df['damageDealt'].quantile(0.75)
IQR = Q3 - Q1
upper = df['damageDealt'] >= (Q3+1.5*IQR)
lower = df['damageDealt'] <= (Q1-1.5*IQR)
a = set(np.where(upper)[0])
b = set(np.where(lower)[0])
a = a.union(b)
for i in a:
    df['damageDealt'][i] = median
```

```python
df.plot.scatter(x = 'damageDealt' , y = 'winPlacePerc')
plt.show()
```

After removing the outliers, we can see the scatter plots are filled. That is a person can win/lose irrespective of the damage

```python
df.plot.scatter(x = 'kills' , y = 'winPlacePerc')
plt.show()
```

```python
df.plot.hexbin(x = 'kills' , y = 'winPlacePerc')
plt.show()
```

Let us remove outliers and try again

```python
median = df['kills'].median()
Q1 = df['kills'].quantile(0.25)
Q3 = df['kills'].quantile(0.75)
IQR = Q3 - Q1
upper = df['kills'] >= (Q3+1.5*IQR)
lower = df['kills'] <= (Q1-1.5*IQR)
a = set(np.where(upper)[0])
b = set(np.where(lower)[0])
a = a.union(b)
for i in a:
    df['kills'][i] = median
```

```python
df.plot.scatter(x = 'kills' , y = 'winPlacePerc')
plt.show()
```
